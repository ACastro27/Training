#### 1- Load pandas and read the file 'example.xlsx'

```python
import pandas as pd

data = pd.read_excel('example.xlsx')
data
```

#### 2 - Find out what are the data types of each column

```python
data.info()
```

#### 3 - Prepare the data of the file so there's only numeric values (superior to 1) in area and price

```python
data = data[(data['area'] > 1) & (data['price'] > 1)]
data
```

#### 4 - Display all the apartments in Nagüeles with more than 100 of area, less than 300 and a price inferior to 200.000, assign these to a new variable

```python
newdata = data[(data['type'] == 'apartments') & (data['location name'] == 'Nagüeles')]
interest = newdata[(newdata['area'] > 100) & (newdata['area'] < 300) & (newdata['price'] < 200000) & (newdata['price'] > 1)]
interest
```

#### 5 - Use the previous variable, create a new column in the dataframe named 'price/m' with the prices per square meter, then sort by it and reorder the index

```python
interest['price/m'] = interest['price']/interest['area']

interest = interest.sort_values(by=['price/m']).reset_index()

interest = interest.drop('index', axis=1)

interest
```

#### 6 - Still using the previous variable, remove the apartments that have a price/m > 2500 and then drop the column type

```python
interest = interest[interest['price/m'] <= 2500].drop('type',axis=1)
interest
```

#### Second Resolution (with a loop, automatic)

```python
tri = data[(data['area'] > 1) & (data['price'] > 1)]
tri['price/m'] = data['price']/data['area']
tri = tri.loc[:, ['location name','type','price','area','price/m']].dropna()
tri
```

```python
for name, group in tri.groupby(['type','location name']):
    if name[0] == 'apartments':
        group[(group['area'] > 100) &
              (group['area'] < 300) &
              (group['price'] < 200000)].to_excel(name[1]+'.xlsx',index=False) #we're saving directly to excel
```
